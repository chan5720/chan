```python
import math
import matplotlib.pyplot as plt
class cos_graph:
  def generate_x_y(self):
    self.x=[]
    self.y=[]
    self.dx=(self.end-self.st)/(self.n-1)
    for i in range(self.n):
      x_=self.st+self.dx*i
      self.x.append(x_)
      self.y.append(math.cos(x_))
  def plot(self):
    plt.plot(self.x,self.y)
    plt.show()
```

```python
g1=cos_graph()
g1.st=0
g1.end=4*math.pi
g1.n=100
g1.generate_x_y()
g1.plot()
```

```python
class TextProcessor:
    def processLine(self, line, wordCounts):
        line = self.replacePunctuations(line)  # Replace punctuation with spaces
        words = line.split()  # Get words from a line
        for word in words:
            if word in wordCounts:
                wordCounts[word] += 1
            else:
                wordCounts[word] = 1

    def replacePunctuations(self, line):
        for ch in line:
            if ch in ".,;:":
                line = line.replace(ch, " ")
        return line
```

```python
def wordCount(self, filename):
  f = open("I_have_a_dream.txt","r")
  lines = f.readlines()
  wordCounts = {}
  for line in lines:
    self.processLine(line.lower(), wordCounts)
  return wordCounts
```

```python
def arrange_pairs(self, wordCounts):
  pairs = list(wordCounts.items())
  items = [[x,y] for (y,x) in pairs]
  items.sort()
  items = {x:y for [y,x] in items}
  return items
```
